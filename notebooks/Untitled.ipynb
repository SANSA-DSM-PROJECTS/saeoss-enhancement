```python
import os
import glob
from osgeo import gdal

# Get all .tif files in the ../data/ directory
tif_files = glob.glob('../data/*.tif')

# Loop through each file and open it with GDAL
for file_path in tif_files:
    ds = gdal.Open(file_path)
    if ds is None:
        print(f"Failed to open: {file_path}")
    else:
        print(f"Successfully opened: {file_path}")
        print("[ RASTER BAND COUNT ]: ", ds.RasterCount)
        for band in range(ds.RasterCount ):
            band += 1
            print("[ GETTING BAND ]: ", band)
            srcband = ds.GetRasterBand(band)
            if srcband is None:
                continue
        
            stats = srcband.GetStatistics( True, True )
            if stats is None:
                continue

            print("[ STATS ] =  Minimum=%.3f, Maximum=%.3f, Mean=%.3f, StdDev=%.3f" % ( stats[0], stats[1], stats[2], stats[3] ))
```
